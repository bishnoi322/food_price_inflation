```python
food.set_index('date', inplace = True)
```

```python
annual_inflation = food.groupby('country')['Inflation'].resample('Y').mean()
```

```python
annual_inflation = annual_inflation.unstack(level=0)
```

```python
import matplotlib.pyplot as plt
```

```python
annual_inflation.plot(legend=True)
plt.xlabel('Year')
plt.ylabel('Average Food Price Inflation')
plt.title('Food Price Inflation Trends Over the Past Decade')
plt.legend(bbox_to_anchor=(1.05, 1),
                         loc='upper left', borderaxespad=0.)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(food['country'], food['Inflation'])
plt.xlabel('Country')
plt.ylabel('Mean Inflation Rate (2013-2022)')
plt.title('Overall Food Price Inflation Trend Over the Past Decade')
plt.xticks(rotation=90)  # Rotate x-axis labels for readability
plt.tight_layout()
```
